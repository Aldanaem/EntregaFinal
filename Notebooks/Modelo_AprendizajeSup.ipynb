***5. Modelo KNN***

En primera instancia se elige como modelo de clasificacin KNN (K-Nearest Neighbors) ya que contamos con datos etiquetados y buscamos predecir una variable target.

***5.1. Primera Prueba***

Se busca a travs del modelo KNN sin la variable (['EquipajeExtra']) desarrollar la perfomance mas eficiente para desarrollar nuestro algoritmo de maching learning.

```python
knn = KNeighborsClassifier()
```

* Entrenamos el modelo

```python
knn.fit(X_train, y_train)
```

* Predecimos

```python
predicciones_cla = knn.predict(X_test)
```

* Graficamos

```python
plt.figure()
random_confusion = confusion_matrix(y_test, predicciones_cla)
sns.heatmap(random_confusion, cmap="YlGnBu",annot=True);
```

* Mtricas

Para analizar el rendimiento del modelo se calculan las siguientes mtricas: La Accuracy es la proporcin de predicciones correctas (verdaderos positivos ms verdaderos negativos) en comparacin con el total de predicciones. 
La precisin se centra en las predicciones positivas y mide la proporcin de verdaderos positivos con respecto al total de predicciones positivas (verdaderos positivos ms falsos positivos).
El recall se centra en las instancias positivas y mide la proporcin de verdaderos positivos con respecto al total de instancias positivas (verdaderos positivos ms falsos negativos). 
El F1 score es la media armnica de precision y recall. 

```python
print(f"Accuracy: {accuracy_score(y_test, predicciones_cla)*100 :.2f}%")
print(f"Precision: {precision_score(y_test, predicciones_cla)*100 :.2f}%")
print(f"Recall: {recall_score(y_test, predicciones_cla)*100 :.2f}%")
print(f"F1 score: {f1_score(y_test, predicciones_cla)*100 :.2f}%")
```

Los resultados sugieren que el modelo tiene un rendimiento moderado. La Accuracy es relativamente baja, la precisin del 72.04% indica que cuando el modelo predice la clase positiva, tiene una tasa razonable de aciertos. El recall del 80.57% sugiere que el modelo es efectivo para capturar la mayora de las instancias positivas en el conjunto de datos. El F1 score del 76.07% proporciona una medida combinada de precisin y recall

***6. Optimizacin de los parmetros del clasificador KNN con Gridsearch***

Se buscan diferentes valores para el nmero de vecinos (n) y se evalua el rendimiento del modelo KNN en conjuntos de entrenamiento y prueba. Buscamos encontrar el nmero ptimo de vecinos que maximiza el rendimiento del modelo.

* Target X

```python
ACC_train = []
ACC_test = []

vecinos = [1, 3, 5, 10, 15, 20, 27, 50,55,59,67]
for n in vecinos:

    clf = KNeighborsClassifier(n_neighbors=n)

    clf.fit(X_train, y_train)
    y_train_pred = clf.predict(X_train)

    train_acc = accuracy_score(y_train, y_train_pred)
    ACC_train.append(train_acc)

    y_test_pred = clf.predict(X_test)
    test_acc = accuracy_score(y_test, y_test_pred)
    ACC_test.append(test_acc)
```

El siguiente grfico nos puede ayudar a identificar si hay un nmero de vecinos que optimice la presicin del modelo. 

```python
plt.plot(vecinos,ACC_train,'o-',label='train', color='#DCE775')
plt.plot(vecinos,ACC_test,'o-',label='test', color='#EC407A')
plt.legend()
sns.despine();
```

Muestra cmo vara la Accuracy en los conjuntos de entrenamiento y prueba a medida que cambia el nmero de vecinos

* Establecemos los parmetros

```python
param_grid_vec = {
    'n_neighbors': [18, 20, 22],
    'weights': ['uniform','distance'],
    'metric': ['euclidean', 'manhattan'],
}
```

* Entrenamos le modelo

```python
knn_search = KNeighborsClassifier()
```

```python
vecinos_search = GridSearchCV(estimator=knn_search, param_grid=param_grid_vec,
                         cv=3, verbose=2, n_jobs=-1)
```

```python
vecinos_search.fit(X_train, y_train)
```
